```python
import math
class calculator():
    def __init__(self, input1, input2):
        self.input1=input1
        self.input2=input2
    
    def addition(self):
        return self.input1+self.input2
    def subtraction(self):
        return self.input1-self.input2
    def multiplication(self):
        return self.input1*self.input2
    def division(self):
        return self.input1/self.input2
    
    
    def square(self):
        return self.input1**2
    def cube(self):
        return self.input1**3
    def sqrt(self):
        return math.sqrt(self.input1)
    def log(self):
        return math.log(self.input1)
    def exponentiate(self):
        return self.input1**self.input2
    def floor_division(self):
        return self.input1//self.input2
    def modulo(self):
        return self.input1%self.input2
    def sin(self):
        return math.sin(self.input1)
    def cos(self):
        return math.cos(self.input1)
    def tan(self):
        return math.tan(self.input1)
    def factorial(self):
        return math.factorial(self.input1)
    
    
input_1=int(input("Enter first number: "))
input_2=int(input("Enter second number: "))
my_calc=calculator(input_1, input_2)

choice=1
while choice!=0:
    print()
    print("0. Exit")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Square (1st input)")
    print("6. Cube (1st input)")
    print("7. Square root (1st input)")
    print("8. Logarithm (1st input)")
    print("9. Exponentiate")
    print("10. Floor division")
    print("11. Modulo")
    print("12. Sine (1st input)")
    print("13. Cosine (1st input)")
    print("14. Tangent (1st input)")
    print("15. Factorial (1st input)")
    
    choice=int(input("Enter your choice.... "))
    
    if choice==1:
        print(f"Result is: {my_calc.addition()}")
    elif choice==2:
        print(f"Result is: {my_calc.subtraction()}")
    elif choice==3:
        print(f"Result is: {my_calc.multiplication()}")
    elif choice==4:
        print(f"Result is: {my_calc.division()}")
    elif choice==5:
        print(f"Result is: {my_calc.square()}")
    elif choice==6:
        print(f"Result is: {my_calc.cube()}")
    elif choice==7:
        print(f"Result is: {my_calc.sqrt()}")
    elif choice==8:
        print(f"Result is: {my_calc.log()}")
    elif choice==9:
        print(f"Result is: {my_calc.exponentiate()}")
    elif choice==10:
        print(f"Result is: {my_calc.floor_division()}")
    elif choice==11:
        print(f"Result is: {my_calc.modulo()}")
    elif choice==12:
        print(f"Result is: {my_calc.sin()}")
    elif choice==13:
        print(f"Result is: {my_calc.cos()}")
    elif choice==14:
        print(f"Result is: {my_calc.tan()}")
    elif choice==15:
        print(f"Result is: {my_calc.factorial()}")
    elif choice==0:
            print("Exit")
    else:
            print("Sorry invalid choice")
print()
    
```
